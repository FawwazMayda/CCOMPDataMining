```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
idx=['ID', 'LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6', 'default']
```

```python
print(idx)
```

```python
df.head()
```

```python
df.columns=idx
df.head()
```

```python
pd.crosstab(df['EDUCATION'],df.default).plot(kind='bar')
plt.show()
```

```python
to_keep=['LIMIT_BAL','SEX','EDUCATION','MARRIAGE','AGE','PAY_1','PAY_2','PAY_6','BILL_AMT1']
X=df.loc[:,to_keep]
Y=df.loc[:,'default']
print(X.shape)
print(Y.shape)
print(X.head())
print(Y.head())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
rf=RandomForestClassifier(n_estimators=20)
dt=DecisionTreeClassifier()
```

```python
a100=pd.read_csv('AUTO100.csv')
print(a100.head())
y_only=a100['Y'].values
```

Import test Data

```python
test_data=pd.read_csv('Data Asli/test.csv')
test_col=idx
test_data.columns=test_col
test_data.head()
```

```python
x_test=test_data.loc[:,to_keep]
x_test.head()
```

```python
rf.fit(X,Y)
dt.fit(X,Y)
y_pred_rf =rf.predict(x_test)
y_pred_dt=dt.predict(x_test)
y_true_rf = y_pred_rf == y_only
y_true_dt = y_pred_dt == y_only
print(len(y_true_rf[y_true_rf==True])/9000)
print(len(y_true_dt[y_true_dt==True])/9000)
```

```python
for i in range(25,50):
    rf=RandomForestClassifier(n_estimators=i)
    rf.fit(X,Y)
    y_pred_rf =rf.predict(x_test)
    y_true_rf = y_pred_rf == y_only
    print(str(i)+"-->>>",end="")
    print(len(y_true_rf[y_true_rf==True])/9000)
```

```python
for i in range(5,50):
    dt=DecisionTreeClassifier(random_state=i)
    dt.fit(X,Y)
    y_pred_dt =dt.predict(x_test)
    y_true_dt = y_pred_dt == y_only
    print(str(i)+"-->>>",end="")
    print(len(y_true_dt[y_true_dt==True])/9000)
```

```python
for i in range(5,50):
    dt=DecisionTreeClassifier(max_depth=i)
    dt.fit(X,Y)
    y_pred_dt =dt.predict(x_test)
    y_true_dt = y_pred_dt == y_only
    print(str(i)+"-->>>",end="")
    print(len(y_true_dt[y_true_dt==True])/9000)
```

```python
for i in range(5,50):
    dt=DecisionTreeClassifier(min_samples_split=i)
    dt.fit(X,Y)
    y_pred_dt =dt.predict(x_test)
    y_true_dt = y_pred_dt == y_only
    print(str(i)+"-->>>",end="")
    print(len(y_true_dt[y_true_dt==True])/9000)
```

```python
for i in range(5,50):
    dt=DecisionTreeClassifier(min_samples_leaf=i)
    dt.fit(X,Y)
    y_pred_dt =dt.predict(x_test)
    y_true_dt = y_pred_dt == y_only
    print(str(i)+"-->>>",end="")
    print(len(y_true_dt[y_true_dt==True])/9000)
```

```python
mg, i_t, j_t, k_t, l_t, maxs = 0, 0, 0, 0, 0, 0

for i in range(15,25):
    for j in range(2, 5):
        for k in range(2, 5):
            for l in range(15, 25):
                print(i, j, k, l, end = '')
                dt=DecisionTreeClassifier(max_depth = i, min_samples_split = j, min_samples_leaf=k, max_leaf_nodes = l)
                dt.fit(X,Y)
                y_pred_dt =dt.predict(x_test)
                y_true_dt = y_pred_dt == y_only
                print(" --> ",end="")
                mg = len(y_true_dt[y_true_dt==True])/9000
                if mg > maxs :
                    i_t, j_t, k_t, l_t, maxs = i, j, k, l, mg
                print(len(y_true_dt[y_true_dt==True])/9000)
            print()
        print()
    print()
    
print(i_t, j_t, k_t, l_t, maxs)
```
